## ¿Qué es la IA?

Es una de las ciencias más recientes que abarca una gran variedad de áreas: desde el aprendizaje y la percepción, a otras más específicas como el diagnóstico de enfermedades

En términos generales, la IA no sólo intenta comprender, también se esfuerza en construir entidades inteligentes

### Subdisciplinas

La IA tiene varias subdisciplinas, como lo son machine learning o aprendizaje automático, deep learning o aprendizaje profundo, IA discriminativa y LLM's junto ccon IA Generativa

Machine learning se refiere al paradigma de la computación donde la capacidad para resolver el problema dado se construye a partir de ejemplos anteriores

Deep learning se refiere al entrenamiento de redes neuronales más profundas que las entrenadas habitualmente

La IA generativa genera contenidos de forma automática en respuesta a instrucciones escritas en interfaces conversacionales de lenguaje natural (prompts)

### Aprendizaje automático

En el aprendizaje automático hay subdisciplinas también como el aprendiaje supervisado, no supervisado y por refuerzo

El aprendizaje supervisado se enfoca en tareas de clasificación y regresión, donde el modelo aprende a predecir etiquetas (clasificación) o valores continuos (regresión) a partir de datos de entrada

El aprendizaje no supervisado se enfoca en descubrir patrones en datos no etiquetados, como la reducción de dimensiones para enfrentar la maldición de la dimensionalidad y el clustering o agrupamiento para identificar grupos de datos similares

El aprendizaje por refuerzo se basa en la analogía de prueba y error, donde un agente interactúa con un entorno y aprende a maximizar una recompensa a través de la toma de decisiones secuenciales, utilizando estrategias denominadas políticas

## ¿Qué es la IA débil y fuerte?

#### IA débil

La IA débil, o comúnmente llamada IA estrecha, se centra en realizar una tarea específica dentro de un dominio limitado, como responder preguntas basadas en la entrada del usuario o jugar ajedrez

Este tipo de IA no busca replicar la inteligencia humana en su totalidad

Los automóviles autónomos y los asistentes virtuales, son ejemplos de IA débil

#### IA fuerte

La IA fuerte, o comúnmente llamada IA general, es aquella que, si puediera desarrollarse, poseería una inteligencia y autoconciencia iguales a la de los humanos, además de una capacidad de resolver una gama limitada de problemas

Tiene como objetivo crear máquinas inteligentes que sean indistinguibles de la mente humana. Pero, tiene que aprender a través de entradas y experiencias, progresando constantemente y mejorando sus habilidades con el tiempo

No existen ejemplos claros de IA fuerte, pero la IA está innovando rápidamente, como la teoría de superinteligencia artificial, que, sigue siendo especulativa porque no se han logrado ejemplos que superen a la IA fuerte en inteligencia y capacidad humanas

#### IA débil vs IA fuerte

La IA débil puede realizar solo un tipo de tarea, no puede realizar dos al mismo tiempo. Mientras que la IA fuerte puede realizar diversas funciones y, finalmente, aprender a resolver nuevos problemas

La IA débil se basa en la interferencia humana para definir los parámetros de sus algoritmos de aprendizaje y proporcionar los datos de entrenamiento relevantes para garantizar la precisión. Mientras que la entrada humana acelera la fase de crecimiento de la IA fuerte, que no es necesaria, pero con el timepo desarrolla una conciencia similar a la humana en lugar de simularla como la IA débil

## Enfoques de IA e inteligencia computacional

#### IA clásica para reglas lógicas

La IA clásica se basa en reglas, es decir, es una IA programada que tiene una estructura definida pero no aprende. Los sistemas expertos, donde se programan una serie de preguntas para dar un diagnóstico, son un ejemplo de este enfoque.

Si el sistema se diseñó para determinar la toma de un medicamente, una regla podría indicar:

si se tiene cierto nivel de glucosa, y los síntomas A y B, ENTONCES, se debe recomendar tomar X fármaco en Y dosis

Así, a medida que se sigue investigando, se añaden más reglas al sistema, que se tendrán que actualizar a la base de conocimiento para reglejar la información sobre los nuevos descubrimientos

#### Redes neuronales para aprendizaje de patrones

Las redes neuronales aprenden de los datos y no requieren que un experto proporcione las reglas. Las redes neuronales artificiales forman parte del aprendizaje automático y analizan grandes volúmenes ded atos, extraen estadísticas y clasifican los resultados

Como su aprendizaje se asemeja más a la inteligencia humana por su estructura en capas que le permite aprender patrones complejos a partir de datos, son adecuadas para abordar con éxito problemas como la clasificación de imágenes y la traducción de idiomas

#### Evolutivo con algoritmos genéticos para optimización

La inteligencia evolutiva, o la computación evolutiva, simula la evolución natural, la que está basada en el principio de la supervivencia del más apto propuesta por Darwin, para resolver problemas complejos de búsqueda, optimización y aprendizaje

La idea de los algoritmos evolutivos es el usar mecanismos como la reproducción, mutación y selección natural para generar y mejorar soluciones candidatas a problemas de optimización que se adapten mejor a las metas deseadas a lo largo del tiempo

#### Inteligencia computacional

La inteligencia computacional, por otra parte, se centra en el diseño de sistemas informáticos inteligenten que imiten tanto la naturaleza cmo el razonamiento lingüístico humano para resolver problemas complejos

En este enfoque se pueden utilizar otros enfoques para el desarrollo de este tipo de sistemas inteligentes:

*   emulación del razonamiento lingüístico humano a través de la teoría de conjuntos difusos
*   imitación del sistema nervioso a través de las redes neuronales
*   algoritmos de optimización bioinspirados




Esta inteligencia computacional, a través de esta combinación de enfoques, permite adaptarse, aprender y generalizar frente a entornos complejos, cambiantes o inciertos, ofreciendo soluciones más flexibles y robustos que métoros o técncias tradicionales

## Sistemas inteligentes

Los sistemas inteligentes manifiestan un comportamiento inteligente para resolver problemas complejos. Además, emula aspectos exhibidos por el ser humano ante la naturaleza y utiliza técnicas de IA para percibir, razonar y actuar de manera autónoma

Estos sistemas pueden adoptar diversas formas, desde robots hasta programas de software que pueden ser aplicados en una amplia gama de campos o áreas

Algunos ejemplos son:



*   Netflix como servicio de recomendación
*   Amazon sugiriendo productos
*   Chatbots de atención al cliente
