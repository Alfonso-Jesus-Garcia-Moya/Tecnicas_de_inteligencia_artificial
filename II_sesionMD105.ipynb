## Variables

En Python, una variable es un espacio de memoria que se reserva para almacenar información. De esta forma, cada que se usa el nombre de la variable en el programa, realmente se está refiriendo al valor del dato almacenado en la memoria.

Por ejemplo, se escribe un nombre para la variable, se usa el signo igual y se le asigna un valor:

```python
x = 13
print(x)
```

En el ejemplo, se utiliza el bloque de la memoria para almacenar el número 13. Después se le asigna a este bloque el nombre de x. A partir de entonces, cada que se utilice la palabra x, en realidad se estará refiriendo al valor almacenado dentro del bloque de momemoria, al cual se apodó como X.

Para nombrar variables existen estas reglas que se recomienda seguir:



*   empezar con una letra o un guión bajo
*   no debe comenzar con ningún número
*   distingue mayúsculas de minúsculas
*   no debe contener espacios ni caracteres especiales
*   hay palabras reservadas que no pueden usarse porque Python las usa para otra funcionalidad
*   es mejor elegir nombres descriptivos, que abreviaciones

## Operadores lógicos

*   Si se cumplen las dos condiciones (  𝑎∧𝑏  ): a and b
*   Si se cumple alguna de las condiciones (  𝑎∨𝑏  ): a or b
*   Si no se cumple esa condición (  𝑎⎯⎯⎯  ): not







```python
x = int(input("Por favor, ingresa un número:"))
y = x > 7 and x < 13
z = x > 7 or x < 13
k = not x > 7

print("¿x > 7 y x < 13?", y)
print("¿x > 7 ó x < 13?", z)
print("¿Es x NO mayor a 7?", k)
```
