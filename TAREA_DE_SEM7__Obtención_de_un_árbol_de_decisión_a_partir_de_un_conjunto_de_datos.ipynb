```python
# importa la librería Pandas;
import pandas as pd
import matplotlib.pyplot as plt

# importa el conjunto de datos breast_cancer;
from sklearn.datasets import load_breast_cancer

# importa el módulo train_test_split;
from sklearn.model_selection import train_test_split

# importa el clasificador;
from sklearn.tree import DecisionTreeClassifier, plot_tree

# explora el conjunto;
df = load_breast_cancer()

# muestra los nombres de las columnas, así como sus tipos de datos;
df_features = pd.DataFrame(df.data, columns=df.feature_names)
df_features.info()

# calcula la estadística básica del conjunto de datos;
print(df_features.describe())

# divide el conjunto en X y y;
X, y = df.data, df.target

# divide el conjunto en entrenamiento y prueba;
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# construye el modelo con el clasificador árboles de decisión;
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# obtén el árbol de decisión;
fig = plt.figure(figsize=(25, 15))
plot_tree(clf,
          feature_names=df.feature_names,
          class_names=df.target_names,
          filled=True,
          rounded=True,
          fontsize=10)
plt.show()

# obtén el desempeño de la métrica accuracy.
print(clf.score(X_test, y_test))
```
