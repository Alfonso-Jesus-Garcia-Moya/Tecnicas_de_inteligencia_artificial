### Inteligencia computacional

Ejemplo con un perceptrón (neurona artificial)

El perceptrón fue la primera neurona propuesta que se caracteriza por ser el modelo más básico de una red neuronal

Consiste en un conjunto de pesos, valores de entrada o puntuaciones y un umbral. Siendo así, el primer paso para aprender aprendizaje automático y profundo

Un perceptrón de una sola capa analiza los objetos linealmente separables con resultados binarios


Este modelo consta de una sola neurona con entradas, pesos y una función de activación

Componente básico de las redes neuronales artificiales que aprende a partir de datos etiquetados para realizar tareas de clasificación o reconocimiento de patrones, normalmente datos linealmente separables

##### Representación gráfica

image.png

Un perceptron consta de:

* entradas: el perceptrón toma varias entradas (x1, x2, ... , xn)
* ponderación: cada entrada está asociada a una ponderación (w1, w2, ... , wn)
* sesgo: añade un término de sesgo (b) para desplazar el límite de decisión
* función de activación: el perceptron utiliza una función escalonada (función de umbral simple) para determinar si la suma ponderada de las entradas más el sesgo está por encima o por debajo de un umbral determinado
* salida binaria: la salida de un perceptron es binaria (1 o 0), lo que lo hace adecuado para problemas de clasificación linealmente separable

La representación matemática es:

image.png

##### Ventajas

* simplicidad: es fácil de entender y de implementar, convirtiéndolo en una excelente opción para cuando se inicia en el campo de machine learning
* rápido entrenamiento: debido a su estructura simple, el entrenamiento del perceptron es generalmente rápido, especialmente en problemas de clasificación lineal
* entrenamiento supervisado: el perceptron es un modelo supervisado, lo que significa que aprende a partir de ejemplos etiquetados, lo cual es ideal para problemas en los que se dispone de datos etiquetados

##### Desventajas/limitaciones

* separabilidad lineal: solo puede resolver problemas que sean linealmente separables, es decir, solo lo que se puede clasificar en datos que puedan separarse mediante una línea recta o un hiperplano
* modelo de una sola capa: el perceptron original es un modelo de una sola capa y no tiene capas ocultas, lo que limita su capacidad de expresión

Se utilizará el perceptrón en una máquina que debe aprender la tabla de verdad del operador AND

image.png

```python
import numpy as np

#entradas (X1, X2)
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

#pesos y sesgo, previamente ajustados para representar el operador AND
w = np.array([0.5, 0.5])  #pesos
b = -0.7                  #sesgo

#función de activación
def perceptron(x):
    z = np.dot(x, w) + b
    return 1 if z >= 0 else 0

#probando combinaciones
for i in range(len(X)):
    print(f"Entrada {X[i]} → Predicción: {perceptron(X[i])}")
```



*   La máquina recibe dos entradas (0 o 1)
*   Multiplica esas entradas por sus pesos (importancia de cada entrada)
*   Suma un sesgo y aplica función de decisión
*   Al final, logra la función lógica AND sin que se programe un if explícitio


¿Qué hace el sistema?

No hay reglas fijas como en el sistema experto básico, pero sí razonamiento aproximado

El perceptrón no tenía reglas escritas como en un if

Aprendió solo a partir de ejemplos, esto es aprendizaje adaptativo, lo que converge a inteligencia computacional

```python
#datos de entrada: (x1, x2)
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])


#salida esperada
y = np.array([0, 0, 0, 1])


#inicializando pesos y sesgo
w = np.random.rand(2)   #pesos aleatorios
b = np.random.rand(1)   #sesgo


#tasa de aprendizaje
lr = 0.1

#entrenando perceptrón
for epoch in range(20):
    for i in range(len(X)):
        #predicoendo
        z = np.dot(X[i], w) + b
        pred = 1 if z >= 0.5 else 0


        #actualizando los pesos de acuerdo al error de predicción
        error = y[i] - pred
        w += lr * error * X[i]
        b += lr * error


print("Pesos aprendidos:", w)
print("Sesgo aprendido:", b)

#probando el perceptrón
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    pred = 1 if z >= 0.5 else 0
    print(f"Entrada {X[i]} → Predicción: {pred}")
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
```

```python
#cargando el conjunto de datos
df = load_breast_cancer()
```

```python
#dividiendo en x/y
X, y = df.data, df.target
```

```python
#dividiendo los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#creando clasificador
clf = Perceptron(random_state=42)
clf.fit(X_train, y_train)
```

```python
#evaluando con accuracy
accuracy = clf.score(X_test, y_test)
print("Exactitud del Perceptron:", accuracy)
```

```python
#evaluando con precision
precision = clf.score(X_test, y_test)
print("Precision del Perceptron:", precision)
```
